```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
```

```python
top_20_countries = df["Country"].value_counts().head(20)
```

```python
plt.figure(figsize=(15, 6))
for i, value in enumerate(top_20_countries.values):
  plt.text(i, value + 0.2, str(value), ha='center', va='bottom', fontsize=12, color='black')
top_20_countries.plot(kind="bar")
plt.xticks(rotation=45, ha="right")
plt.xlabel("Country")
plt.ylabel("Number of respondents")
plt.title("Top 20 countries with the highest number of respondents")
```

```python
age = df["Age"].value_counts()
```

```python
for i, value in enumerate(age.values):
  plt.text(i, value + 0.2, str(value), ha='center', va='bottom', fontsize=12, color='black')

age.plot(kind="bar")
plt.xlabel("Age")
plt.ylabel("Number of respondents")
plt.title("Age distribution of developers")
```

```python
years_code = df["YearsCode"].value_counts()
```

```python
plt.figure(figsize=(30, 6))
for i, value in enumerate(years_code.values):
  plt.text(i, value +0.2, str(value), ha="center", va="bottom", fontsize=12, color="black")

years_code.plot(kind="bar")
plt.xlabel("Years of Coding")
plt.ylabel("Number of respondents")
plt.title("Years of Coding")
plt.tight_layout()
```
